# BFS approach

```python
class State():
    #creating a state
    def __init__(self, cl, ml, b, cr, mr):
        self.cl=cl
        self.ml=ml
        self.cr=cr
        self.mr=mr
        self.b=b
        self.par=''

    def goal(self): #for checking whether the goal is reached
        if self.cl == 0 and self.ml == 0: return True
        return False

    def valid(self): #for checking whether the state is valid
        if self.ml >= 0 and self.mr >= 0 and self.cl >= 0 and self.cr >= 0 :
            if(self.ml == 0 or self.ml >= self.cl) and (self.mr == 0 or self.mr >= self.cr): 
                return True
        return False
    
    def __eq__(self, other):
        return self.cl == other.cl and self.ml == other.ml and self.cr == other.cr and self.mr == other.mr and self.b == other.b
    
    def __hash__(self):
        return hash((self.cl, self.ml, self.cr, self.mr, self.b))

def successors(cstate):
    children = []
    if cstate.b == 'r':
        nstate = State(cstate.cl, cstate.ml + 2, 'l', cstate.cr, cstate.mr - 2)
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl + 2, cstate.ml, 'l', cstate.cr - 2, cstate.mr)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl + 1, cstate.ml + 1, 'l', cstate.cr - 1, cstate.mr - 1)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl, cstate.ml + 1, 'l', cstate.cr, cstate.mr - 1)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl + 1, cstate.ml, 'l', cstate.cr - 1, cstate.mr)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        
    else:
        nstate = State(cstate.cl, cstate.ml - 2, 'r', cstate.cr, cstate.mr + 2)
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl - 2, cstate.ml, 'r',cstate.cr + 2, cstate.mr)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl - 1, cstate.ml - 1, 'r', cstate.cr + 1, cstate.mr + 1)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl, cstate.ml - 1, 'r', cstate.cr, cstate.mr + 1)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl - 1, cstate.ml, 'r', cstate.cr + 1, cstate.mr)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
    return children

def breadth_first_search():
    initial_state = State(3, 3, 'l', 0, 0)
    if initial_state.goal(): return initial_state
    recordx = list()
    visited = set()
    recordx.append(initial_state)
    while recordx:
        state = recordx.pop(0)
        if state.goal(): return state
        visited.add(state)
        children = successors(state)
        for i in children:
            if (i not in visited) or (i not in recordx): recordx.append(i)
    return ''


def display(solution):
    path = []
    path.append(solution)
    par = solution.par
    while par:
        path.append(par)
        par = par.par
    for i in range( len(path)):
        state = path[len(path) - i - 1]
        j=1 if state.b=='l' else 0
        if i == len(path) - 1:
            print(f'goal state: ({state.cl},{state.ml},0 : {state.cr},{state.mr},1)')
            print(f'it took {i} steps to achieve goal state')
        else:print(f'State {i}: ({state.cl},{state.ml},{abs(j)} : {state.cr},{state.mr},{abs(1-j)})')

if __name__ == "__main__":
    display(breadth_first_search())
```

# DFS approach

```python
class State():
    #creating a state
    def __init__(self, cl, ml, b, cr, mr):
        self.cl=cl
        self.ml=ml
        self.cr=cr
        self.mr=mr
        self.b=b
        self.par=''

    def goal(self): #for checking whether the goal is reached
        if self.cl == 0 and self.ml == 0: return True
        return False

    def valid(self): #for checking whether the state is valid
        if self.ml >= 0 and self.mr >= 0 and self.cl >= 0 and self.cr >= 0 :
            if(self.ml == 0 or self.ml >= self.cl) and (self.mr == 0 or self.mr >= self.cr): 
                return True
        return False
    
    def __eq__(self, other):
        return self.cl == other.cl and self.ml == other.ml and self.cr == other.cr and self.mr == other.mr and self.b == other.b
    
    def __hash__(self):
        return hash((self.cl, self.ml, self.cr, self.mr, self.b))

def successors(node):
    #to find all the successors of the state
    children = []
    if node.b == 'r': #when boat is on right bank
        temstate = State(node.cl, node.ml + 2, 'l', node.cr, node.mr - 2)
        if temstate.valid():
            temstate.par = node
            children.append(temstate)
            
        temstate = State(node.cl + 2, node.ml, 'l', node.cr - 2, node.mr)
        if temstate.valid():
            temstate.par = node
            children.append(temstate)
            
        temstate = State(node.cl + 1, node.ml + 1, 'l', node.cr - 1, node.mr - 1)
        if temstate.valid():
            temstate.par = node
            children.append(temstate)
            
        temstate = State(node.cl, node.ml + 1, 'l', node.cr, node.mr - 1)
        if temstate.valid():
            temstate.par = node
            children.append(temstate)
            
        temstate = State(node.cl + 1, node.ml, 'l', node.cr - 1, node.mr)
        if temstate.valid():
            temstate.par = node
            children.append(temstate)
        
    else: #when boat is on left bank
        temstate = State(node.cl, node.ml - 2, 'r', node.cr, node.mr + 2)
        if temstate.valid():
            temstate.par = node
            children.append(temstate)
            
        temstate = State(node.cl - 2, node.ml, 'r',node.cr + 2, node.mr)
        if temstate.valid():
            temstate.par = node
            children.append(temstate)
            
        temstate = State(node.cl - 1, node.ml - 1, 'r', node.cr + 1, node.mr + 1)
        if temstate.valid():
            temstate.par = node
            children.append(temstate)
            
        temstate = State(node.cl, node.ml - 1, 'r', node.cr, node.mr + 1)
        if temstate.valid():
            temstate.par = node
            children.append(temstate)
            
        temstate = State(node.cl - 1, node.ml, 'r', node.cr + 1, node.mr)
        if temstate.valid():
            temstate.par = node
            children.append(temstate)
    return children

def dfs():
    recordx = State(3, 3, 'l', 0, 0)
    if recordx.goal(): return recordx
    openlist = [recordx] #openlist is a list of states
    closed = [] #closed is a list of visited states
    while openlist:
        state = openlist.pop(0)
        if state.goal(): return state
        closed.append(state)
        children = successors(state)
        temp=[]
        for i in children:
            if i not in closed: 
                temp.append(i) 
        openlist = temp[::-1] + openlist #adding children to openlist
    return ''


def display(solution):
    path = [solution]
    par = solution.par #for parent tracking
    while par:
        path.append(par)
        par = par.par
    for i in path[::-1]: #printing in reverse order from initial state to final
        j=1 if i.b=='l' else 0
        print(f'({i.cl},{i.ml},{abs(j)} : {i.cr},{i.mr},{abs(1-j)})')
    print(f'it took {len(path)-1} steps to reach goal state')
    
if __name__ == "__main__":
    display(dfs())
```
