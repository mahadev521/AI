# DFID approach

```python
'''
nstate -> New state
cstate -> current state
cl -> cannibal left
cr -> cannibal RIGHT
l -> left
r -> RIGHT
b -> boat
'''

class State():
    def __init__(self, cl, ml, b, cr, mr):
        self.cl,self.ml,self.b,self.cr,self.mr,self.par = cl,ml,b,cr,mr,''

    def goal(self):
        if self.cl == 0 and self.ml == 0: return True
        return False

    def valid(self):
        if self.ml >= 0 and self.mr >= 0 and self.cl >= 0 and self.cr >= 0 and (self.ml == 0 or self.ml >= self.cl) and (self.mr == 0 or self.mr >= self.cr): return True
        return False


def successors(cstate):
    children = []
    if cstate.b == 'r':
        nstate = State(cstate.cl, cstate.ml + 2, 'l', cstate.cr, cstate.mr - 2)
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl + 2, cstate.ml, 'l', cstate.cr - 2, cstate.mr)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl + 1, cstate.ml + 1, 'l', cstate.cr - 1, cstate.mr - 1)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl, cstate.ml + 1, 'l', cstate.cr, cstate.mr - 1)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl + 1, cstate.ml, 'l', cstate.cr - 1, cstate.mr)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        
    else:
        nstate = State(cstate.cl, cstate.ml - 2, 'r', cstate.cr, cstate.mr + 2)
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl - 2, cstate.ml, 'r',cstate.cr + 2, cstate.mr)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl - 1, cstate.ml - 1, 'r', cstate.cr + 1, cstate.mr + 1)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl, cstate.ml - 1, 'r', cstate.cr, cstate.mr + 1)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl - 1, cstate.ml, 'r', cstate.cr + 1, cstate.mr)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
    return children


def depth_limited_search(node, depth):
    if node.goal(): return node
    elif depth == 0: return None
    else:
        children = successors(node)
        for child in children:
            result = depth_limited_search(child, depth - 1)
            if result:
                return result
        return None


def iterative_deepening_depth_first_search(root, max_depth):
    for i in range(max_depth):
        return depth_limited_search(root, max_depth)


def display(solution):
    path = list()
    path.append(solution)
    par = solution.par
    while par:
        path.append(par)
        par = par.par
    print("(cannibal (l), missionaries (l), b(l): cannibals (r), missionaries (r), b(r))")
    print("state 0: (3,3,1 : 0,0,0)")
    for i in range(1, len(path)):
        state = path[len(path) - i - 1]
        j=1 if state.b=='l' else 0
        if i == len(path) - 1:
            print(f'goal state: ({state.cl},{state.ml},0 : {state.cr},{state.mr},1)')
            print(f'it took {i} steps to achieve goal state')
        else:print(f'State {i}: ({state.cl},{state.ml},{abs(j)} : {state.cr},{state.mr},{abs(1-j)})')

if __name__ == "__main__":
    initial_state = State(3, 3, 'l', 0, 0)
    solution = iterative_deepening_depth_first_search(initial_state, 11)
    display(solution)
```

# BFS approach

```python
'''
nstate -> New state
cstate -> current state
cl -> cannibal left
cr -> cannibal RIGHT
l -> left
r -> RIGHT
b -> boat
'''
class State():
    def __init__(self, cl, ml, b, cr, mr):
        self.cl,self.ml,self.b,self.cr,self.mr,self.par = cl,ml,b,cr,mr,''

    def goal(self):
        if self.cl == 0 and self.ml == 0: return True
        return False

    def valid(self):
        if self.ml >= 0 and self.mr >= 0 and self.cl >= 0 and self.cr >= 0 and (self.ml == 0 or self.ml >= self.cl) and (self.mr == 0 or self.mr >= self.cr): return True
        return False

def successors(cstate):
    children = []
    if cstate.b == 'r':
        nstate = State(cstate.cl, cstate.ml + 2, 'l', cstate.cr, cstate.mr - 2)
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl + 2, cstate.ml, 'l', cstate.cr - 2, cstate.mr)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl + 1, cstate.ml + 1, 'l', cstate.cr - 1, cstate.mr - 1)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl, cstate.ml + 1, 'l', cstate.cr, cstate.mr - 1)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl + 1, cstate.ml, 'l', cstate.cr - 1, cstate.mr)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        
    else:
        nstate = State(cstate.cl, cstate.ml - 2, 'r', cstate.cr, cstate.mr + 2)
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl - 2, cstate.ml, 'r',cstate.cr + 2, cstate.mr)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl - 1, cstate.ml - 1, 'r', cstate.cr + 1, cstate.mr + 1)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl, cstate.ml - 1, 'r', cstate.cr, cstate.mr + 1)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
        nstate = State(cstate.cl - 1, cstate.ml, 'r', cstate.cr + 1, cstate.mr)
        
        if nstate.valid():
            nstate.par = cstate
            children.append(nstate)
    return children

def breadth_first_search():
    initial_state = State(3, 3, 'l', 0, 0)
    if initial_state.goal(): return initial_state
    recordx = list()
    visited = set()
    recordx.append(initial_state)
    while recordx:
        state = recordx.pop(0)
        if state.goal(): return state
        visited.add(state)
        children = successors(state)
        for i in children:
            if (i not in visited) or (i not in recordx): recordx.append(i)
    return ''


def display(solution):
    path = []
    path.append(solution)
    par = solution.par
    while par:
        path.append(par)
        par = par.par
    for i in range( len(path)):
        state = path[len(path) - i - 1]
        j=1 if state.b=='l' else 0
        if i == len(path) - 1:
            print(f'goal state: ({state.cl},{state.ml},0 : {state.cr},{state.mr},1)')
            print(f'it took {i} steps to achieve goal state')
        else:print(f'State {i}: ({state.cl},{state.ml},{abs(j)} : {state.cr},{state.mr},{abs(1-j)})')

if __name__ == "__main__":
    display(breadth_first_search())
```
